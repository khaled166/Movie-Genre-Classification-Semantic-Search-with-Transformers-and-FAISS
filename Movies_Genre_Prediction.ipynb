```python
def impute_value(main_val, aux_val):
    if pd.isna(main_val) or str(main_val).strip().lower() in ["", "unknown", "nan"]:
        return aux_val
    return main_val
```

```python
def parse_genre_string(genre_string):
    # Convert to string and lower it
    genre_string = str(genre_string).lower()
    # Remove square brackets and extra whitespace
    genre_string = genre_string.strip("[]")
    # Split on slashes and spaces
    genre_list = re.split(r'[/,\s]+', genre_string)
    # Remove empty strings and clean genre names
    genre_list = [re.sub(r'[^a-z]', '', g) for g in genre_list if g]
    return genre_list

# Apply to Genre column
df['Genre'] = df['Genre'].fillna('').swifter.apply(parse_genre_string)
df['Genre']
```

```python
def remove_outliers_iqr(df, column, min_limit=None, max_limit=None, iqr_range=1.5, verbose=True):
    """
    Remove outliers based on token length of a text column using IQR or manual limits.
    Always removes rows with length < min_limit or > max_limit if provided.
    """
    df_copy = df.copy()

    # Step 1: Calculate text length
    length_col = f"{column}_Length"
    df_copy[length_col] = df_copy[column].astype(str).apply(lambda x: len(x.split()))

    # Step 2: Compute IQR bounds only if manual bounds not provided
    Q1 = df_copy[length_col].quantile(0.25)
    Q3 = df_copy[length_col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - iqr_range * IQR if min_limit is None else min_limit
    upper_bound = Q3 + iqr_range * IQR if max_limit is None else max_limit

    # Step 3: Apply filtering
    filtered_df = df_copy[
        (df_copy[length_col] >= lower_bound) & (df_copy[length_col] <= upper_bound)
    ].copy()

    if verbose:
        print(f" Original dataset: {len(df)} rows")
        print(f" Cleaned dataset:  {len(filtered_df)} rows")
        print(f" Removed {len(df) - len(filtered_df)} rows outside range {int(lower_bound)}‚Äì{int(upper_bound)}")

    return filtered_df
```

#### ‚úÖ Why Use a Pipeline?

Creating a pipeline ensures consistent preprocessing and modeling steps. It prevents data leakage, simplifies cross-validation, and allows for efficient deployment in production environments.

---

#### üîß Pipeline Workflow Covers:

- **Multi-Label Target Handling**
  - The target column `'Genre'` contains multiple labels per sample.
  - Used `MultiLabelBinarizer` to convert it into a binary matrix representation.

- **Text Feature Engineering**
  - Applied `TfidfVectorizer` on the `'Plot'` column to transform raw movie descriptions into numerical embeddings that preserve important narrative signals.

- **Categorical Feature Encoding**
  - Employed `Onehotencoder` to convert non-numeric columns (like `'Origin/Ethnicity'`) into machine-readable form.

- **Feature Scaling**
  - Used `RobustScaler(with_centering=False)` to scale numerical features and reduce the influence of outliers.

- **Handling Imbalanced Data**
  - Addressed imbalance using models with `class_weight='balanced'` rather than synthetic oversampling, which yielded better performance in experimentation.

- **Model Selection for Multi-Label Classification**
  - Utilized `OneVsRestClassifier` to extend classic classifiers to multi-label problems.
  - Evaluated multiple models inside the pipeline:
    - Logistic Regression
    - Decision Tree
    - Linear Support Vector
    - LightGBM
    - Random Forest

---

#### üìä Evaluation Metrics

- **Cross-Validation**
  - Used 10-fold cross-validation to assess the generalization performance of each model.
  
- **Metrics Used**
  - `Hamming Loss`: Measures how many labels are incorrectly predicted.
---

> üîÅ The use of `Pipeline` made the modeling process modular, reliable, and suitable for real-world deployment where consistent preprocessing is critical.


```python
from nltk.tokenize import word_tokenize
numeric_cols   = ['Release Year', 'Decade']
categorical    = ['Origin/Ethnicity']
text_cols      = ['Title', 'Plot', 'Cast', 'Director']
preprocess = ColumnTransformer(
    transformers=[
        #('num', 'passthrough', numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical),
        ('txt_plot',     TfidfVectorizer(max_features=4000, ngram_range=(1,2),
        tokenizer=word_tokenize,stop_words='english'), 'Plot'),
        ('txt_cast',     TfidfVectorizer(max_features=1500), 'Cast'),
        ('txt_director', TfidfVectorizer(max_features=2000), 'Director')
    ],
    remainder='drop'
)
```

<center><span style="font-size:30px; color:White; font-weight:bold;">Non-Spoiler Summaries for Movie Plots</span></center>

#### üß† Goal:
To generate brief, non-spoiler summaries of movie plots that avoid revealing major twists or endings, making them safe for readers who haven't watched the films.

---

#### üîç Approach:

- **LLM Used:**  
  Leveraged the `google/flan-t5-xl` transformer model via Hugging Face's `pipeline("text2text-generation/summarization")` to produce high-quality summaries.

- **Custom Prompting:**  
  Each movie plot was summarized using a tailored prompt to instruct the model *not* to reveal spoilers:  
  *"Summarize the following movie plot without revealing major twists or the ending:"*

- **Sample Size:**  
  Applied the summarization to a small random sample (e.g., 10 rows) from the `Plot` column for demonstration and evaluation.

---

#### üß™ Evaluation:

- **Metric Used:**  
  BERTScore (F1) was computed to evaluate the semantic similarity between the generated summaries and the original movie plots.

- **Observations:**  
  - BERTScore assesses contextual similarity rather than exact word overlap, making it more suitable for abstractive summaries.

  -  The average BERTScore F1 value indicated strong semantic preservation in the summaries, despite omitting spoilers and compressing details.

  -   This validates that the summarizer retained key plot essence while avoiding sensitive content.
---

> üöÄ This step adds explainability and usability to the dataset, especially for user-facing applications like movie recommendation systems or content browsing tools.


```python
from transformers import pipeline

# Initialize the instruction-following model
summarizer = pipeline("text2text-generation", model="google/flan-t5-xl")

# Sample non-empty plots with titles
sample_df = df[df['Plot'].str.len() > 700][['Title', 'Plot']].sample(n=3, random_state=42).reset_index(drop=True)

# Define the instruction-based summarizer
def summarize_non_spoiler(plot_text):
    prompt = (
    "Summarize the following movie plot without revealing major twists or the ending:\n\n" + plot_text[:3000]
)
    
    # Print the prompt for debugging
    print("---- Prompt Sent to Model ----")
    print(prompt[:500])
    print("------------------------------\n")
    
    try:
        response = summarizer(prompt, max_length=100, min_length=30, do_sample=False)
        return response[0]['generated_text'] if response else "[No summary returned]"
    except Exception as e:
        return f"[Error: {e}]"

# Apply summarization
results = []
for idx, row in sample_df.iterrows():
    summary = summarize_non_spoiler(row['Plot'])
    results.append({
        'Title': row['Title'],
        'Original Plot': row['Plot'],
        'Non-Spoiler Summary': summary
    })

# Create DataFrame with titles as index
summary_df = pd.DataFrame(results).set_index("Title")

summary_df
```

<center><span style="font-size:30px; color:White; font-weight:bold;">Vector DB & RAG</span></center>

#### üß† Goal:
To implement a semantic search and retrieval-augmented generation (RAG) system that enhances information access by returning contextually relevant movie plots and answers to user queries.

---

#### üì¶ Components:

- **Sentence Embedding Model:**  
  Used `sentence-transformers/all-MiniLM-L6-v2` to convert movie plots into dense vector embeddings for semantic understanding.

- **Vector Database:**  
  Built using **FAISS (Facebook AI Similarity Search)** to store and index plot embeddings efficiently and support fast nearest-neighbor retrieval.

---

#### üîç Semantic Search Functionality:

- Encoded all plot texts into normalized vectors.
- Used cosine similarity via `IndexFlatIP` in FAISS to retrieve the top `k` most relevant plots for a given query.
- Created a `search_movies(query)` function to return movie titles, plots, and relevance scores.

---

#### üß© Retrieval-Augmented Generation (RAG):

- Implemented a lightweight RAG system using `google/flan-t5-base`.
- For any user question, the system:
  1. Retrieves top-matching plots from the FAISS vector DB.
  2. Concatenates them into a prompt passed to the LLM.
  3. Generates a contextual answer based on retrieved knowledge.

---

#### ‚öôÔ∏è Sample Use Case:

- **Query:** *"what is the shawshank redemption movie about"*  
- **System Output:**  
  - **RAG Answer:** Synthesized based on retrieved movie plots.  
  - **Retrieved Context:** Titles and summaries of the top related films.

---

> üí° This setup makes the system capable of intelligent movie recommendation, plot-based search, and natural-language Q&A ‚Äî ideal for enhancing user interactivity in movie exploration tools.


```python
 # Build enriched text + FAISS index  (re-run only if df changes)
#assert {'Title', 'Plot'}.issubset(df.columns), "df needs Title & Plot columns"

# Create richer document text for better semantic retrieval
def make_rag_text(row):
    # If Genre is list -> join; else cast to str
    genre_str = ", ".join(row['Genre']) if isinstance(row['Genre'], list) else str(row.get('Genre', ''))
    return f"Title: {row['Title']}\nGenre: {genre_str}\nPlot: {row['Plot']}"

df['RAG_Text'] = df.swifter.apply(make_rag_text, axis=1)
```

```python
df.head(3)  # Display the first 3 rows of the DataFrame with RAG_Text
```

```python
# -- Encode
encoder   = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

embeddings = encoder.encode(
    df['RAG_Text'].tolist(),
    show_progress_bar=True,
    convert_to_numpy=True,
    normalize_embeddings=True 
)


# -- FAISS cosine index
dim   = embeddings.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(embeddings)

print(f"‚úì FAISS vector DB built ({index.ntotal} docs, dim={dim})")
```

```python
def search_movies(query: str, top_k: int = 5) -> pd.DataFrame:
    q_vec          = encoder.encode([query], normalize_embeddings=True)
    scores, idx    = index.search(q_vec, top_k)
    hits           = df.iloc[idx[0]].copy()
    hits['Score']  = scores[0]
    return hits[['Title', 'Plot', 'Score']].reset_index(drop=True)
```

```python
gen_name   = "google/flan-t5-base"          # small, fast, open-source
tokenizer  = AutoTokenizer.from_pretrained(gen_name) # tokenizer for text2text generation
generator  = AutoModelForSeq2SeqLM.from_pretrained(gen_name) # load the model for text2text generation
gen_pipe   = pipeline("text2text-generation", model=generator, tokenizer=tokenizer) # text2text generation pipeline
```

```python
def rag_qa(question: str, top_k: int = 5, max_new_tokens: int = 128):
    retrieved = search_movies(question, top_k=top_k)
    
    # Assemble context with title & plot for each hit
    context = "\n\n---\n\n".join(
        f"Title: {row.Title}\nPlot: {row.Plot}" for _, row in retrieved.iterrows()
    )

    prompt = (
        "Using the information in the movie plots below, answer the question "
        "WITHOUT revealing major spoilers.\n\n"
        f"{context}\n\n"
        f"Question: {question}\nAnswer:"
    )

    answer = gen_pipe(prompt, do_sample=False, max_new_tokens=max_new_tokens)[0]['generated_text']
    return answer.strip(), retrieved[['Title', 'Score']]

# üîé Demo
user_q = "what is the shawshank redemption movie about"
ans, ctx = rag_qa(user_q, top_k=5)

print("Question:", user_q)
print("\nAnswer:", ans)
print("\nTop-k retrieved:")
display(ctx)
```

<center><span style="font-size:30px; color:White; font-weight:bold;">Q&A Deployment Using Gradio</span></center>

#### üéØ Goal:
Deploy an interactive interface that allows users to ask natural language questions about movies and receive intelligent, context-aware answers powered by RAG (Retrieval-Augmented Generation).

---

#### üöÄ Tools Used:

- **Gradio:**  
  A fast and user-friendly library to create web-based UIs for ML models.
  
- **Vector DB (FAISS):**  
  Performs semantic retrieval of movie plots relevant to the user‚Äôs question.

- **Language Model:**  
  `google/flan-t5-base` generates concise and informative answers based on the retrieved context.

---

#### üß© How It Works:

1. **User Inputs a Question** ‚Äì e.g., *‚Äúwhat is the shawshank redemption movie about?‚Äù*
2. **Plot Retrieval** ‚Äì Top 3 most relevant movie plots are fetched from the FAISS vector database using semantic similarity.
3. **Answer Generation** ‚Äì The retrieved content is passed into the LLM along with the question to generate a grounded response.
4. **Answer Display** ‚Äì Gradio interface presents both the generated answer and a preview of retrieved movie titles/plots.

---

#### üñ• Interface Features:

- **Text Input Box** for user questions.
- **Answer Section** showing generated output.
- **Context Preview** listing relevant movie titles retrieved from the database.
- Fast response time and intuitive UX, all run locally or hosted online via Gradio sharing.

---

> ‚úÖ This deployment enables end users to interact with your movie intelligence system in real-time ‚Äî transforming raw movie data into actionable, conversational insights.


```python
import gradio as gr # Gradio for web interface

def rag_gradio(query): 

    """
    Gradio interface function for RAG Q&A.
    Takes a user query, retrieves relevant movie plots, and generates an answer.
    Args:
        query (str): User's question about a movie. 
    Returns:
        tuple: A tuple containing:
            - movie_titles (str): Formatted string of top-k movie titles and their scores.
            - answer (str): Generated answer to the user's query.
    """
    
    answer, hits = rag_qa(query)
    movie_titles = "\n".join(f"- {row['Title']} ({row['Score']:.2f})" for _, row in hits.iterrows()) # Format titles with scores
    return movie_titles, answer # Return titles and answer

gr.Interface(
    fn=rag_gradio, # Function to call for RAG Q&A
    inputs=gr.Textbox(label="Ask about a movie"),# User input textbox for movie questions
    outputs=[
        gr.Textbox(label="Top Matches"), # Output textbox for top-k movie titles and scores
        gr.Textbox(label="Answer") # Output textbox for the generated answer
    ],
    title="üé¨ RAG Movie Q&A",
    description="Ask about movie plots. Powered by FAISS + Flan-T5.",
).launch()
```
